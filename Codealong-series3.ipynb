```python
# import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# declare url to be scrapped
url = "https://www.scrapethissite.com/pages/simple/"
```

```python
# use requests to make a HTTP GET request
res = requests.get(url)
res
```

```python
# check the type of object res is
type(res)
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text)
soup
```

```python
# find a div with the class name of "container"
container = soup.find("div", class_ = "container")
container
```

```python
# find a div with the class name of "container"
container = soup.find_all("div", class_ = "container")
container
```

```python
#assign the countries to variable countries
countries = container[1]
countries
```

```python
country_rows = countries.find_all("div", class_= "row")
country_rows
```

```python
country_rows = country_rows[3:]
country_rows
```

```python
for x in country_rows:
    print(x.find_all('h3'))
    print(" ")
```

```python
country_list = []

#loop through all the rows
for x in country_rows:
    # loop through all the individual box in the row
    for y in x.find_all('h3'):
        # print the h3 tag with stripped the whitespaces
        country_list.append(y.text.strip())

country_list
```

```python
capital_list = []

#loop through all the rows
for x in country_rows:
    # loop through all the individual box in the row
    for y in x.find_all('span', class_ = "country-capital"):
        # print the h3 tag with stripped the whitespaces
        capital_list.append(y.text.strip())
        
capital_list
```

```python
population_list = []

#loop through all the rows
for x in country_rows:
    # loop through all the individual box in the row
    for y in x.find_all('span', class_ = "country-population"):
        # print the h3 tag with stripped the whitespaces
        population_list.append(int(y.text.strip()))
        
population_list
```

```python
area_list = []

#loop through all the rows
for x in country_rows:
    # loop through all the individual box in the row
    for y in x.find_all('span', class_ = "country-area"):
        # print the h3 tag with stripped the whitespaces
        area_list.append(float(y.text.strip()))

area_list
```

```python
data_dict = {"country": country_list,
             "capital": capital_list,
             "population": population_list,
             "area": area_list}

df = pd.DataFrame(data_dict)
df
```
